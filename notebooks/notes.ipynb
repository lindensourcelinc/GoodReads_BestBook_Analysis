Histogram to display distribution of ratings across all the titles in the dataset

```python
book_rates = up_df.dropna(subset=['rating'])

plt.figure(figsize=(12,8))

sns.histplot(data=book_rates, x='rating', bins=20, kde=True, color= 'green')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.title("GoodReads Ratings", fontsize=24)
plt.xlabel('Ratings (0 to 5)', fontsize=16)
plt.ylabel('Number of Books', fontsize=16)



plt.tight_layout
plt.show()
```

Line chart to display any possible correlation between book length in pages and reader ratings.

```python
up_df['pages'] = pd.to_numeric(up_df['pages'], errors='coerce')
up_df['pages']

pages_df = up_df.dropna(subset=['pages', 'rating'])
pages_df
```

```python
up_df['page_bins'] = pd.cut(up_df['pages'], bins=range(0, 2000, 100 ), labels=range(100, 2000, 100 ))

pages_df = up_df.groupby('page_bins', as_index=False)['rating'].mean()


plt.figure(figsize=(10,6))
sns.lineplot(data=pages_df, x='page_bins', y='rating', marker='o', color="#791717", linewidth=3)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.title('Is There Any Correlation Between Book Length and Rating?', fontsize=18)
plt.xlabel('Book Length in Pages (by 100s)', fontsize=12)
plt.ylabel('Book Ratings (0-5)', fontsize=12)

plt.xticks(range(0, 2100, 100), rotation=45)

plt.tight_layout()
plt.show()
```
