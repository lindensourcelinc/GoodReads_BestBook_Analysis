Updated publishDate column to datetime data type.

The publish dates are not all formatted the same and in string format. 
Column reformatted to be recognizable to pandas so dates could be coverted to datetime.


```python
# reset the column to restore data 
up_df['publishDate'] = df['publishDate']

# removed all the extra fluff from the publishDate column that read as a string to be recognized as dates and then converted to a datetime data type

clean_dates = up_df['publishDate'].astype(str).str.replace(r'(\d+)(st|nd|rd|th)', r'\1', regex=True)

# converted publishDate column to datetime data type, format='mixed' because of the varying date formats in the column
up_df['publishDate'] = pd.to_datetime(clean_dates, format='mixed', errors='coerce')

up_df['publishDate']
```

Price column changed from string to float data type.
    All missing values filled with "NA".

```python
# converted price column to float
up_df['price'] = pd.to_numeric(up_df['price'], errors='coerce')

# filled missing values with NA
up_df['price'] = up_df['price'].fillna("NA")

up_df['price']
```

```python
# Trimmed the chart for cleaner visualization
high_rates = up_df[up_df['rating'] > 4.0]

author_title_count = high_rates.groupby('author')['title'].nunique().sort_values(ascending=False)


plt.figure(figsize=(10,8))
plt.bar(author_title_count.head(10).index, author_title_count.head(10).values,
        color = "#493236",
        width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.xlabel('Author')
plt.ylabel('Title Count')
plt.title("Top 10 Authors with Books Rated 4.0 and Above in GoodReads ")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

Noted the above chart reflects authors with the most books being considered for BBE rated 4.0 and above, not the highest rated authors. 
    Will rework the code.

```python
# Calculated the average ratings and the book count of each author to determine an author statistic to measure by

author_stats = up_df.groupby('author', as_index=False).agg(
    avg_rating=('rating', 'mean'),
    book_count=('rating', 'count')
)

# Used a minimum book count to deter skewed results of authors with one 5.0-rated book outranking an author with more well-rated books

min_books = 3
top_authors = author_stats[author_stats['book_count'] >= min_books]

top_10 = top_authors.sort_values(by='avg_rating', ascending=False).head(10)

# Horizontal chart chosen to better display author names
plt.figure(figsize=(10,6))
ax = sns.barplot(data=top_10,
            x='avg_rating',
            y='author',
            color='#175579')

# Coded to show the book count inside the bars
for index, row in top_10.reset_index(drop=True).iterrows():
    ax.text(
        x=row['avg_rating'] - 0.2,
        y=index,
        s=f"{row['book_count']} books",
        color='white',
        va='center',
        ha='right',
        fontweight=14
    )

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.xlabel('Average Book Rating (0-5)', fontsize=11)
plt.ylabel('Author', fontsize=11)

# Centered the title across the whole graphic vs it hugging the right side over the bars; feels more balanced to look at
plt.suptitle("Top 10 Highest-Rated Authors in the GoodReads BBE Running", fontsize=18, ha='center')


# Zoomed in on the x-axis limit since the comparison is so slight
plt.xlim(4.0, 5.0)

plt.tight_layout()
plt.show()

```

What question or idea does this chart help answer?

- Who are the authors of the most well-rated books in the GoodReads Best Book Ever running and how many books of their books are being considered?

Why is this chart type appropriate?

- This chart displays the author names, book counts, all aligned over their average book rating in a visually agreeable manner that allows all the data to be read and interpreted easily. 

What design choices did you make (color, labels, order, scale)?

- All choices made to see the information on the chart easily without overcrowding the graphic.

How did you ensure accuracy and avoid misleading design?

- The book counts are listed, the average rating limits on the x-axis are fair and reasonable, as well set the minimum books at 3 to manage skewed data in the chart.

Is there a possible correlation between book length in pages and reader ratings?

```python
# Converted pages column to integer data type
up_df['pages'] = pd.to_numeric(up_df['pages'], errors='coerce')
up_df['pages']

# Dropped all columns that were missing data points in both pages and rating columns
pages_df = up_df.dropna(subset=['pages', 'rating'])
pages_df
```

```python
up_df['page_bins'] = pd.cut(up_df['pages'], bins=range(0, 2000, 100 ), labels=range(100, 2000, 100 ))

pages_df = up_df.groupby('page_bins', as_index=False)['rating'].mean()


plt.figure(figsize=(10,6))
sns.lineplot(data=pages_df, x='page_bins', y='rating', marker='o', color="#0B3935FF", linewidth=3)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.title('Is There Any Correlation Between Book Length and Rating?', fontsize=18)
plt.xlabel('Book Length in Pages (by 100s)', fontsize=12)
plt.ylabel('Book Ratings (0-5)', fontsize=12)

plt.xticks(range(0, 2100, 100),rotation=45)

plt.tight_layout()
plt.show()
```

What question or idea does this chart help answer?

- If there's a correlation between book length and reader ratings? 
This could possibly help publishers and authors determine choosing books and parameters for best sales projections according to consumer preferences, maybe even prompt further look into the longer length books that rated higher and some possible similarities that could spell out what makes for a best-seller.

Why is this chart type appropriate?

- This chart clearly displays the upward trending relationship between higher rated books and longer page lengths. 


What design choices did you make (color, labels, order, scale)?

- Pages are measured by 100s so there's clarity when aligning the plot points with the page lengths.
- The ratings start at zero, but you can still clearly see even with many of the books being rated 4.0 and above, which is reasonable considering they were all in the running for the best book ever, there's still a solid distinction of higher ratings for longer books.
- Color was chosen as neutral, but high contrast option against the stark white background.


How did you ensure accuracy and avoid misleading design?

- No persuasive language used. 
- All the labels and units of measure are accurately portrayed with no biased skewing. 
- The data is allowed to speak for itself.
